```python
import librosa
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import random
```

```python
train_csv = pd.read_csv('100ms_training_data.csv')
test_csv = pd.read_csv('100ms_test_data.csv')
guitar_df = train_csv[train_csv['instrument'] == 'Sound_Guitar']
piano_df = train_csv[train_csv['instrument'] == 'Sound_Drum']
drum_df = train_csv[train_csv['instrument'] == 'Sound_Violin']
violin_df = train_csv[train_csv['instrument'] == 'Sound_Piano']
```

```python
fig,axs=plt.subplots(4,6, figsize=(25,15))

for i in range(6):
    index = random.randint(0,len(guitar_df))
    melspec = np.loadtxt(f"./100ms_train_mel_spec/{guitar_df['mel spec ref'].iloc[index]}").reshape((128, 44))
    pic = librosa.display.specshow(melspec, sr=22050, x_axis='time', y_axis='mel', ax=axs[0][i])
    axs[0][i].set_title('guitar', loc='left')
    
for i in range(6):
    index = random.randint(0,len(piano_df))
    melspec = np.loadtxt(f"./100ms_train_mel_spec/{piano_df['mel spec ref'].iloc[index]}").reshape((128, 44))
    pic = librosa.display.specshow(melspec, sr=22050, x_axis='time', y_axis='mel', ax=axs[1][i])
    axs[1][i].set_title('piano', loc='left')
    
for i in range(6):
    index = random.randint(0,len(violin_df))
    melspec = np.loadtxt(f"./100ms_train_mel_spec/{violin_df['mel spec ref'].iloc[index]}").reshape((128, 44))
    pic = librosa.display.specshow(melspec, sr=22050, x_axis='time', y_axis='mel', ax=axs[2][i])
    axs[2][i].set_title('violin', loc='left')
    
for i in range(6):
    index = random.randint(0,len(drum_df))
    melspec = np.loadtxt(f"./100ms_train_mel_spec/{drum_df['mel spec ref'].iloc[index]}").reshape((128, 44))
    pic = librosa.display.specshow(melspec, sr=22050, x_axis='time', y_axis='mel', ax=axs[3][i])
    axs[3][i].set_title('drum', loc='left')


plt.show()
```
